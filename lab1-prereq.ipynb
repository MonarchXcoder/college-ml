```python
#9
string='01-JUN-2021'
s1=string.split('-')
print('date:',s1[0],' month:',s1[1],' year:',s1[2])
month_dict = {
    "JAN": 1,
    "FEB": 2,
    "MAR": 3,
    "APR": 4,
    "MAY": 5,
    "JUN": 6,
    "JUL": 7,
    "AUG": 8,
    "SEP": 9,
    "OCT": 10,
    "NOV": 11,
    "DEC": 12
}
month_num = month_dict.get(s1[1])
print('The numerical value for', s1[1],'is', month_num)
```

```python
#11
import numpy as np
import matplotlib.pyplot as plt
V1=np.random.randint(1, 101, size=100)
V1_sorted = np.sort(V1)

plt.plot(V1_sorted, label="Original Color")
plt.title("Sorted Vector Plot - Default Color")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
plt.plot(V1_sorted, color='red', label="Red Color")  # Change color to red
plt.title("Sorted Vector Plot - Red Color")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
#12
V2=V1**2
plt.plot(V2, color='green', label="Green Color")  # Change color to green
plt.title("Squared Vector Plot - Green Color")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```
